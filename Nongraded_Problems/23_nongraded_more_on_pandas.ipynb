```python
# P1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
names = {'a': [11, np.nan, 13, 14, 15], 'b': [14, 15, 16, 17, 18], 
         'c': [np.nan, 18, np.nan, 10, 11], 'd': [20, np.nan, 22, 9, np.nan]}
df = pd.DataFrame(data=names)
print("DataFrame object:", df, sep = '\n')
print("\nNumber of NaN values:", df.isnull().values.sum(), sep = '\n')
```

```python
# P2
data = np.array([[2,7,5,7,6,9],[1,1,3,4,9,6],[10,7,0,4,2,8]])
df = pd.DataFrame(data, columns=['A','B','C','D','E','F'], index=[1,2,3])
print(df)
```

```python
df.plot.bar() #groups the values in each row 
plt.legend(loc='upper left')
```

```python
df.transpose().plot.bar() #groups the values in each column 
plt.legend()
```
