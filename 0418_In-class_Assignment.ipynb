### Problem 1

```python
# 1 (50%). Please write a function that takes as input two iterators, 
#e.g., iter1 and iter2, and returns another iterator, in which 
#each item is a tuple of paired elements in the two input iterators. 
#The i-th tuple contains the i-th element from each input iterator. 
#The function stops when the shorter input iterator is exhausted. 
#Please do not use lists, dictionaries, or the built-in function zip() in your implementation. 
#In addition, please do not convert iter1 or iter2 into a tuple in the function.

#For example,

#if iter1=iter(range(1,3)) and iter2=iter(range(6,12)), 
#then the function should return an iterator that contains elements, (1, 6) and (2, 7).

#if iter1=iter(('python', 'ruby', 'Java')) and iter2=iter(('apple', 'strawberry')), 
#then the function should return an iterator that contains elements, 
#('python', 'apple') and ('ruby', 'strawberry').

#In addition to the function, please write some codes to test your function.

def iter_zip(iter1, iter2):
    '''
    Returns an iterator where each item is a tuple of the paired elements
    in the two input iterators.
    The i-th tuple contains the i-th element from each input iterator.
    Stops when the shorter iterator is exhausted.
    '''
    while True:                      #Infinite loop until StopIteration exception is raised
        try:
            current = yield (next(iter1),next(iter2)) #makes a generator object with each next element
        except StopIteration:
            break
    return current   #Returns generator object (type of iterator)
        

iter1 = iter(range(1,3))
iter2 = iter(range(6,12))

for element in iter_zip(iter1,iter2):
    print(element, end=' ')

print()

iter1 = iter(('python', 'ruby', 'Java'))
iter2 = iter(('apple', 'strawberry'))

for element in iter_zip(iter1,iter2):
    print(element, end=' ')
```

```python
#2 (50%). Please write a Python program to define a class named Patient. 
#Each Patient object/instance has three attributes: 
#   name (a string), 
#   age (an integer), 
#   and pid (a unique integer), 
#which represent the name, the age, and the patient ID number of the patient, respectively.

#The Patient class needs to define a 
#  class variable to record the total number of Patient instances created and a 
#  class method to get the total number of Patient instances created. 

#In addition, the class should be defined to 
#support access to attribute values for name, age and pid with the dot operation from outside of the class, 
#and also to support changes of attribute values for name and age with the dot operation. 

#The class does not allow a negative age value assignment with the dot operation, 
#i.e., it should raise an error when the client codes try to assign a 
#negative value to the age attribute using the dot operation.

#In addition to the Patient class, please write some client codes to test your class.

class Patient:
    '''
    Represents a Patient object with attributes for name, age, and patient ID.
    Name, age, and pid are readable. Only name and age are writable.
    Also keeps track of how many patient records there are.
    '''
    patient_count = 0  #Initialize patient count
    
    def __init__(self,name,age,pid):
        self._name = name
        self._age = age
        self._pid = pid
        Patient.increment_patient_count()
    
    @property
    def name(self):
        '''Returns the patient name.'''
        return self._name
    
    @property
    def age(self):
        '''Returns the patient age.'''
        return self._age
    
    @property
    def pid(self):
        '''Returns the patient ID.'''
        return self._pid
    
    @name.setter
    def name(self, new_name):
        '''Sets a new name for patient.'''
        self._name = new_name
    
    @age.setter
    def age(self, new_age):
        '''Sets a new age for patient.'''
        if new_age < 0:
            raise ValueError('Age cannot be negative.')  #Raises an exception if try to set neg value for age
        self._age = new_age
    
    @classmethod
    def increment_patient_count(cls):
        '''Increases the number of patients by one to keep track of number of patient instances created.'''
        cls.patient_count += 1
    
    @classmethod
    def get_patient_count(cls):
        '''Returns the number of patients.'''
        return cls.patient_count
    
a = Patient('Abby',13, 5)
b = Patient('Brett',45, 312)

print(f'Name: {a.name}  Age: {a.age}  PID: {a.pid}')
print(f'Name: {b.name}  Age: {b.age}  PID: {b.pid}')
print('\nPatient count = ',Patient.get_patient_count())

c = Patient('Chris',34,276)

print(f'\nNew patient:: Name: {c.name}  Age: {c.age}  PID: {c.pid}')
print('New Patient Count: ',Patient.get_patient_count())

a.name = 'Abigail' #set new name
b.age = 25         #set new age

print('\nAfter name change:')
print(f'Name: {a.name}  Age: {a.age}  PID: {a.pid}')
print('\nAfter age change:')
print(f'Name: {b.name}  Age: {b.age}  PID: {b.pid}')


try:
    print('\nAttempting age change to -13:')
    a.age = -13           #Raise exception for negative age
finally:
    print('\nAttempting pid change:')
    a.pid = 24            #raises error for trying to set new pid
```
