### Problem 1

```python
#P1 (30%). Please write a function that takes as input two positive integers, num and n. 
#The number of digits of num is not smaller than the value of n. 
#The function needs to delete n digits from num such that the resulting new number is the smallest. 
#The function returns the new number; if all the digits are deleted, the function returns 0.

#For example:

## If num = 2030 and n = 1, then the function should return 30, 
## because deleting the leading 2 produces a number 30, which is the smallest.

## If num = 3712136 and n = 3, then the function should return 1136, 
## because deleting digits 3 (the first one), 7 and 2 produces the smallest number, 1136.

## If num = 19 and n = 2, then the function should return 0, because all the digits are deleted.

import itertools

def digits_removed(num,n):
    '''
    Takes two arguments, num and n. Removes n digits from num and returns the smallest possible number.
    If all digits are removed, returns 0.
    '''
    digit_list = list(digit for digit in str(num)) #Convert digits of the number into a list of digits
    newnum_len = len(digit_list)-n #Length of the new number after removing n digitsif num == n:
    
    if newnum_len <= 0: #If num has the same or fewer digits than n, returns 0.
        return 0
    
    comb_list = list(itertools.combinations(digit_list,newnum_len)) 
    #gives you a list of tuples of all possible digit combinations
    
    new_list = []
    for tupl in comb_list:
        tupl_str = ''   # reset string to empty for each tuple in list
        for digit in tupl:
            tupl_str += str(digit) #adds each digit of the tuple into one string
        new_list.append(int(tupl_str)) #adds the string of digits into the new list as an integer
    return min(new_list) #returns the minimum value among all the possible combinations (in order)
    
print('For num = 2030 and n=1... new number =', digits_removed(2030,1))
print('For num = 3712136 and n = 3... new number =', digits_removed(3712136,3))
print('For num = 19 and n = 2... new number =', digits_removed(19,2))
```

```python
#P2 (35%). There is a CSV file, blood_pressure_data.csv*, 
#which contains the blood pressure information for a group of patients. 
#Columns 1, 2, 3, 4, 5, and 6 of the file correspond to 
#1)patient IDs (note that patient IDs are unique), 2)age, 3)gender, 4)initial systolic pressure, 
#5)drug dose used for treatment, and 6)systolic pressure after treatment, respectively. 
#Please download the CSV file onto your computer and do not modify the content of the file.

## Please write a function to read data from the file and print the 
## Top 10 female patients who have the 
## lowest initial systolic pressure (column4) in ascending order on the screen.

## Please write another function to read data from the file and 
## print the Top 10 patients who have the most amount of systolic pressure reduction 
## (Col4 - Col6; higher number means more reduction)
## after treatment in descending order on the screen.

import os

def female_10lowest_initsys():
    '''
    Prints out the top 10 females with the lowest initial systolic pressure in ascending order.
    '''
    with open('blood_pressure_data.csv') as f:
        print('Top 10 females with lowest initial systolic pressure:')
        lines = f.readlines()
        header = lines[0]
        names = header.strip().split(',')
        print('{0:13} {1}'.format(names[0],names[4]))  #Prints out patient ID and initial systolic blood pressure
        dct = {}
        for row in lines[1:]:
            values = row.strip().split(',')
            if values[2] == '0':                  #If patient is female
                dct[values[0]] = values[3]        #add the patient ID and initial systolic into the dictionary
        for k in sorted(dct.keys(),key=lambda x:dct[x])[:10]: #sort by values in ascending order, top 10
            print('{0:13} {1}'.format(k,dct[k]))
        
def top10systolic_reduction():
    with open('blood_pressure_data.csv') as f:
        print('Top 10 patients with greatest systolic pressure reduction:')
        lines = f.readlines()
        header = lines[0]
        names = header.strip().split(',')
        print('{0:13} {1}'.format(names[0],'Reduction in systolic blood pressure (mm Hg)'))
        dct = {}
        for row in lines[1:]:
            values = row.strip().split(',')
            values[3],values[5] = float(values[3]),float(values[5])
            dct[values[0]] = round(values[3]-values[5], 2) #init - post = reduction. Round to 2 decimal places.
        
        #sort by values in descending order, only including top 10
        for k in sorted(dct.keys(),key=lambda x:dct[x], reverse = True)[:10]:
            print('{0:13} {1}'.format(k,dct[k]))
        
female_10lowest_initsys()
print()
top10systolic_reduction()
```

```python
#P3 (35%). There is a JSON file, olympic_medals.json, which contains the 
#number of Summer Olympic medals for 12 Olympic committees or countries. 
#The country names are in the array associated with the key, “Country”. 
#For each country, the numbers of gold, silver and bronze medals are stored 
#in corresponding elements of the array associated with the key, “Medals”, as shown in Figure 1. 
#Data source: https://en.wikipedia.org/wiki/All-time_Olympic_Games_medal_table

#Please download the JSON file onto your computer and do not modify the content of the file. 
#Please write a program to read data from the file and:

## Print the data within the JSON file in a tabular format on the screen: 
## columns 1, 2, 3, and 4 represent country, gold, silver, and bronze medal numbers, respectively. 
## Please print the data in descending order in terms of the number of gold medals.

## Calculate the total number of medals (including gold, silver and bronze) for each country 
## and then write the country names and their corresponding numbers of 
## gold, silver, bronze and total medals into a CSV file. 
## Please write the data in descending order in terms of the number of total medals. 
## The country (1)names, (2)gold, (3)silver, (4)bronze and (5)total medals should correspond to 
## columns 1, 2, 3, 4 and 5 of the CSV file, respectively.

## Can properly handle the exception that the input file, olympic_medals.json, does not exist.

### For this question, please submit your source codes (.py or .ipynb file), 
### screenshots of the printing of your codes and 
### screenshot of the output CSV file.

import json, csv

try:
    with open('olympic_medals.json', 'r') as json_file:
        data = json.load(json_file)
        countries = data['Country']
        medals = data['Medals']
        dct = {}
        for index in range(len(countries)):
            dct[countries[index]] = medals[index]
        print('{0:<15} {1:<8} {2:<8} {3:<8}'.format('Country','Gold','Silver','Bronze')) #print headings
        for k in sorted(dct.keys(),key=lambda x:dct[x][0], reverse = True): #sort dct by gold medals in desc order
            print('{0:<15} {1:<8} {2:<8} {3:<8}'.format(k,dct[k][0],dct[k][1],dct[k][2]))
    
        for item in dct:
            dct[item].append(sum(dct[item]))  ## Add total medals to the end of value list for each dct entry
    
        with open('olympic_medals.csv', 'w') as output_file:
            f_csv = csv.writer(output_file)
            header = ['Country','Gold','Silver','Bronze','Total_Medals']
            f_csv.writerow(header) #write headings for first row
            
            #sort country names in descending order according to total medals
            sorted_countries = sorted(dct.keys(),key=lambda x:dct[x][3], reverse = True)
            tuple_lst = []
            for item in sorted_countries:
                tuple_lst.append((item,*dct[item]))  #create a list of tuples combining countries and medals
            f_csv.writerows(tuple_lst)        #each subsequent row is a tuple that includes country and medals
except FileNotFoundError:
    print('File not found. Make sure file is in current working directory.')
```
