```python
#P1 (50%). Please write a function that takes a non-empty string, my_str, as input. 
#This string consists of only the following characters: '[', ']', '(', and ')'. 
#The function checks whether the string satisfies the following conditions: 
#(1) each left bracket is closed by a right bracket with the same type, and 
#(2) each pair of left-right brackets appear in the correct order. 
#If the string satisfies these two conditions, the function returns True, otherwise False. 
#In addition to the function, please write some codes to test your function.

def is_matching(my_str):
    '''
    Tells you if your string contains all matching legal brackets
    '''
    def remove_adjacent_matches(string):
        '''
        Removes matching adjacent () and [] pairs.
        '''
        string = string.replace('()','')
        string = string.replace('[]','')
        return string
        
    matching = True
    if my_str.count('(') != my_str.count(')'): #if the number of opening/closing brackets are not the same, 
        matching = False                       #then brackets don't match
    if my_str.count('[') != my_str.count(']'):
        matching = False
    while matching == True and my_str: #while string is not empty and matching remains true
        orig_str = my_str                  #keep resetting orig_str if you remove matching brackets/parentheses
        my_str = remove_adjacent_matches(my_str)    #keep removing adjacent matching brackets or parentheses
        if orig_str == my_str:             #if there were no matching brackets to remove
            matching = False
    return matching


#If my_str = '[]', the function should return True.
#If my_str = '[]()', the function should return True.
#If my_str = '()[]()', the function should return True.
#If my_str = ')(', the function should return False.
#If my_str = '[)', the function should return False.
#If my_str = '[(])', the function should return False.
#If my_str = '()([)]', the function should return False.
#If my_str = '[()]', the function should return True.
#If my_str = '[([])]', the function should return True.

print(is_matching('[]'))
print(is_matching('[]()'))
print(is_matching('()[]()'))
print(is_matching(')('))
print(is_matching('[)'))
print(is_matching('[(])'))
print(is_matching('()([)]'))
print(is_matching('[()]'))
print(is_matching('[([])]'))
```

```python
#P2 (50%). Please write a Python program to count the frequency of each word in a file, document.txt 
#Download document.txt. Please download the file onto your computer and do not modify the content of the file. 
#Your program needs to read the file and

#Print the top 20 words that have the highest frequencies in descending order on the screen. 
#Please print both words and their corresponding frequencies.

#Print the frequency for the word "people" on the screen. 
#Note that "People" is considered a different word from "people".
#Print the frequency for the word "right" on the screen. 
#Note that "rights" or "Right" is considered a different word from "right".
#Punctuation marks are not considered part of a word.

import os, string
#Find absolute path of the file and use that to read the file
f = open(os.path.abspath('document.txt'),'r')

lines = f.readlines()
f.close()  #Remember to close the file

words = []
for line in lines:
    line = line.replace(',',' ') #replace all punctuation with a space
    line = line.replace('.',' ')
    line = line.replace(';',' ')
    line = line.replace('-',' ')
    line = line.replace(':',' ')
    line = line.replace('&',' ')
    words += line.split()        #split all words in each line by spaces and add it to the existing list of words

dct = dict()

for word in words:
    if word in dct:
        dct[word] += 1  #if the word is already in the dictionary, add 1 to the word count
    else:
        dct[word] = 1   #if the word isn't already in the dictionary, start the count at 1

print('Top 20 words:')
print(f"{'Word':10} {'Frequency':4}")
for k in sorted(dct.keys(), key = lambda x: dct[x], reverse = True)[:20]: #sort by values in descending order(first 20 dct entries)
    print(f"{k:10} {dct[k]:4}")

print('Frequency for the word "people": ',dct['people'])
print('Frequency for the word "right": ',dct['right'])
```
